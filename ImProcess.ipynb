```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

def PromPix(M1, M2):
    R = np.zeros([5,5])
    Z = np.zeros([len(M1)])
    for i in range(len(M1)):
        for j in range(len(R)):
            for k in range(len(R)):
                R[j,k] = M2[int(M1[i,1])-2+j, int(M1[i,0])-2+k]
        Z[i] = np.mean(R)
    return Z

#Importar Imagen
img = mpimg.imread('IM1.jpeg')
#Mostrar la Imagen
plt.imshow(img,cmap='gray')
plt.show()
#Seleccionar Puntos
u = np.zeros([2,2])
u[0] = [12.2*len(img)/19., 2.25*len(img.T)/14.25]
u[1] = [16.5*len(img)/19 , 6.75*len(img.T)/14.25]

#Puntos de Recorte de Imagen
u1, u2 = u[0]
v1, v2 = u[1]


#Valor del Pixel
pixelu=img[int(u2), int(u1)]
pixelv=img[int(v2), int(v1)]

#Puntos de Estudio
P = np.zeros([9,2])
for i in range(3):
    for j in range(3):
        P[3*i+j,0] = int(u1+(i+1)*(v1-u1)/4)
        P[3*i+j,1] = int(u2+(j+1)*(v2-u2)/4)
print(P)
#Recorte
plt.clf()
IR = img[int(u2):int(v2),int(u1):int(v1)]
plt.imshow(IR,cmap='gray')
for i in range(9):
    plt.scatter(P[i,0]-u1,P[i,1]-u2, c='none', edgecolor = 'white')
plt.show()
Z = PromPix(P, img)
```
